```python
xlsx = pd.ExcelFile('sample_-_superstore.xls')
```

```python
df_cleaned = klib.data_cleaning(df)
```

```python
df_cleaned.info(memory_usage='deep')
```

```python
df_cleaned.head()
```

```python
df_sales = df_cleaned[["state", "sales"]].copy()
```

```python
df_sales.shape
```

```python
df_sales.info()
```

```python
df_sales.head(50)
```

```python
df_total_sale_by_state = df_sales.groupby("state").sum().reset_index()
```

```python
df_total_sale_by_state.head(50)
```

```python
df_total_sale_by_state ["rank"]= df_total_sale_by_state["sales"].rank(ascending=False)
```

```python
df_total_sale_by_state.shape
```

```python
df_total_sale_by_state.head(50)
```

```python
df_sorted = df_total_sale_by_state.sort_values("rank", ascending=True)
```

```python
df_sorted.shape
```

```python
df_sorted.head(50)
```

```python
df_sorted.info()
```

```python
df_roundup_sales = df_sorted.copy()
```

```python
df_roundup_sales.head(50)
```

```python
import numpy as np
```

```python
df_roundup_sales['sales'] = df_roundup_sales['sales'].apply(np.ceil).astype(int)
```

```python
df_roundup_sales.head(50)
```

```python
df_top_ten_states_by_sales = df_roundup_sales.query("rank <=10") 
```

```python
df_top_ten_states_by_sales.head(11)
```

```python
df_top_ten_states_by_sales["sales"].describe()
```

```python
df_top_ten_states_by_sales.to_csv('topTenStatesBySales.csv', index=False)
```
